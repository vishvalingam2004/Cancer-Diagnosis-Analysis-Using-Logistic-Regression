```python
# Importing Libraries.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("cancer_dataset.csv")
```

```python
df=df.drop(["id","Unnamed: 32"],axis=1)
df.head(7)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
a=["diagnosis"]
encode=OrdinalEncoder()
df[a]=encode.fit_transform(df[a])
```

```python
df
```

```python
df=df.astype({"diagnosis":'int'})
```

```python
sns.countplot(x=df['diagnosis'])
```

```python
sns.countplot(x=df['radius_mean'])
```

```python
cor=df.corr()
cor.style.background_gradient(cmap='autumn')
```

```python
# Train-Test splitting.
x=df[["radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean","compactness_mean","concavity_mean","concave points_mean","symmetry_mean","fractal_dimension_mean","radius_se","texture_se","perimeter_se","area_se","smoothness_se","compactness_se","concavity_se","concave points_se","symmetry_se","fractal_dimension_se","radius_worst","texture_worst","perimeter_worst","area_worst","smoothness_worst","compactness_worst","concavity_worst","concave points_worst","symmetry_worst","fractal_dimension_worst"]]
```

```python
y=df["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=40)
```

```python
# Scaling.
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
x_train=scale.fit_transform(x_train)
x_test=scale.transform(x_test)
```

```python
x_train
```

```python
y_train
```

```python
# Model Building
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
print(ypred)
```

```python
# Accuracy Checking.
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,ypred)
print("Accuracy is:",acc*100)
```

```python
# Evalution Metrics
from sklearn.metrics import confusion_matrix
```

```python
# Confuse Matrix
c=confusion_matrix(y_test,ypred)
print("Confusion Matrix is: \n",c)
```

```python
# Plot confusion matrix
sns.heatmap(c,annot=True,fmt='g',xticklabels=['Malignant','Benign'],yticklabels=['Malignant','Beningn'])
plt.xlabel('Actual',fontsize=13)
plt.ylabel('Prediction',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
# Precision and Recall Checking.
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
pre=precision_score(y_test,ypred)
print("Precision:",pre*100)
rec=recall_score(y_test,ypred)
print("Recall:",rec*100)
F1=f1_score(y_test,ypred)
print("F1 Score:",F1*100)
```
